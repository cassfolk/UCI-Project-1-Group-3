```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt1
import seaborn as sns
```

```python
# Begin -start  ABrar 10/26/2020 - 
# Load data from csv to the data frame 

movie_temp_df = pd.read_csv("Resources/movie_list.csv")
movie_temp_df.head()
```

```python
# Abrar - 10/27/2020 add a column 'Profit'
profit_df = movie_temp_df['revenue'] - movie_temp_df['budget']
#print (profit_df)
movie_temp_df['Profit'] = profit_df 
movie_temp_df
movie_temp_df1 = movie_temp_df
```

```python
# Split up our data into groups based upon 'genres' and summ the Profit column
revenue_grouped_by_Genres = movie_temp_df.groupby('genres')[['Profit']].sum()
print(revenue_grouped_by_Genres)
#revenue_grouped_by_Genres
```

```python
colors = ['red', 'green', 'blue', 'yellow', 'black', 'cyan']

fig, ax = plt.subplots()
ax.bar(revenue_grouped_by_Genres.index, revenue_grouped_by_Genres['Profit'],color = colors, edgecolor = 'green',width = 0.25)

plt.title('Genres vs Profit')
plt.xlabel('Genres')
plt.ylabel('Profit in Millions')
labels = ax.get_xticklabels()
plt.setp(labels, rotation = 45, horizontalalignment = 'right')
plt.ylim(500000, 19900000000)
plt.xlim(0,19) 
plt.subplots_adjust(bottom = 0.3, left = 0.1)
plt.figure( figsize=(10,10))

#plt.figure(figsize=(#,#))
#plt.figure(figsize=(100,100))

    
#matplotlib.pyplot.figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True, FigureClass=<class 'matplotlib.figure.Figure'>, clear=False, **kwargs)[source]

plt.show()
```

```python

sns.set()


sales_data = movie_temp_df1.groupby('genres')[['budget', 'Profit']].sum()

index = np.arange(20)
bar_width = 0.2

r1 = plt1.bar(index, sales_data['budget'],
       bar_width, color = 'green', label = 'Budget')

r2 = plt1.bar(index + bar_width, sales_data['Profit'],
       bar_width, color = 'blue', label = 'Profit')

plt1.xticks(index + bar_width, sales_data.index,
      color = 'green',rotation = 45, horizontalalignment = 'right')

plt1.legend()
plt1.tight_layout()
plt1.show()

# End -Stop  ABrar 10/26/2020 - 
```
