```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt3
import seaborn as sns
```

```python
# Begin -start  ABrar 10/26/2020 - 
# Load data from csv to the data frame 

movie_temp_df = pd.read_csv("Resources/movie_list.csv")
movie_temp_df.head()
```

```python
# Abrar - 10/27/2020 add a column 'Profit'
profit_df = movie_temp_df['revenue'] - movie_temp_df['budget']
#print (profit_df)
movie_temp_df['Profit'] = profit_df 
movie_temp_df
movie_temp_df1 = movie_temp_df
```

```python

sales_data = movie_temp_df1.groupby('genres')[['budget', 'Profit']].sum()

index = np.arange(20)
bar_width = 0.5


plt1.figure( figsize=(12,12))

r1 = plt1.bar(index, sales_data['budget'], bar_width, color = 'green', label = 'Budget')

r2 = plt1.bar(index + bar_width, sales_data['Profit'],bar_width, color = 'blue', label = 'Profit')

plt1.xticks(index + bar_width, sales_data.index,
      color = 'green',rotation = 45, horizontalalignment = 'right')
plt1.title('Genres vs Profit')
plt1.xlabel('Genres')
plt1.ylabel('Profit in Hundred Millions')
plt1.legend()
plt1.tight_layout()
plt1.show()

# End -Stop  ABrar 10/26/2020 - 
```
